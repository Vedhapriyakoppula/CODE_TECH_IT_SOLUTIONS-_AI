```python
encoder = LabelEncoder()
```

```python
data['CGPA_Encoded'] = encoder.fit_transform(data['CGPA'])
```

```python
X_encoded = X_encoded.reshape(-1, 1)
```

```python
X_train, X_test, Y_Depression_train, Y_Depression_test = train_test_split(X_encoded, Y_Depression, test_size=0.2, random_state=30)
X_train, X_test, Y_Work_Pressure_train, Y_Work_Pressure_test = train_test_split(X_encoded, Y_Work_Pressure, test_size=0.2, random_state=30)
```

```python
print(f"Accuracy for Depression: {accuracy_Depression}")
print(f"Accuracy for Work_Pressure: {accuracy_Work_Pressure}")
```
